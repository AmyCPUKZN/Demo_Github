# Cloning a github repo

In this section we will clone an existing github repo.

The command to do this is:

**!git clone https://{git_token}@github.com/{username}/{repository}**,

Replace {git_token} with your token (see step 4 of [this](https://medium.com/analytics-vidhya/how-to-use-google-colab-with-github-via-google-drive-68efb23a42d) Medium article). **Github tokens should be kept safe and secret!**

{username} should be replaced with your Github username as it appears on your Github homepage. 

Finally, replace {repositiry} with the name of the repo you want to clone.



**Only clone once**

```python
!git clone https://{git_token}@github.com/AmyCPUKZN/Demo_Github
```

Check you Google drive for a folder with the same name as your github repo.

```python
%cd Demo_Github
!ls
```

You are now ready to create and edit code and send it to Github!

See step 5.B of the [same](https://medium.com/analytics-vidhya/how-to-use-google-colab-with-github-via-google-drive-68efb23a42d) article.

Any public Github repository can be cloned and anyone else. However, you do not have permission to "push" your edits back to just any repo.

# Important and useful git commands

Cofigure your user name

```python
!git config --global user.email "you@example.com"
!git config --global user.name "Your Name"
```

Use this often to see the status of your local copy of the repo

```python
!git status
```

If you have made in changes you will need to add them

```python
!git add .
```

Once you are happy you can commit your changes

```python
!git commit -m "Write a message here"
```

And "push" them to Github.

```python
!git push
```

In your browser check your Github account for the changes you just made.

```python
# When working in a team you should use "pull" often to check for 
# changes others have made. This avoids merge conflicts.
!git pull
# pull is also useful if you are working on multiple devices to make sure 
# all devices have the most up-to-date code. 
```
